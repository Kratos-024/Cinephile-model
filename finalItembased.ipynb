```python
def make_clean_string(data):
    # Handle non-list input
    if isinstance(data, str):
        # If it's a string that looks like a list (e.g., "['name1', 'name2']"), try to parse it
        if data.startswith("['") and data.endswith("']"):
            try:
                data = eval(data)  # Use eval cautiously
            except:
                return ''
        else:
            data = [data]  # Treat as a single-element list
    elif not isinstance(data, list):
        return ''  # Not iterable, return empty string

    names = []
    for item in data:
        if isinstance(item, str):
            cleaned = item.replace(' ', '').lower()
            names.append(cleaned)

    return ', '.join(names)
```

```python
import ast

def clean_cast_column(cast_value):
    try:
        # If it's already a list, convert to string
        if isinstance(cast_value, list):
            return ', '.join(cast_value)

        # If it's a string, try to parse it as a list
        if isinstance(cast_value, str):
            parsed = ast.literal_eval(cast_value)
            if isinstance(parsed, list):
                return ', '.join(parsed)

        # If it's not list or str, or parsing fails
        return ''

    except (ValueError, SyntaxError):
        return ''  # Return empty string on any parse error
```

```python
best_final_df_with_only_movie_after_2000s['cast'] = best_final_df_with_only_movie_after_2000s['cast'].apply(clean_cast_column)
```

```python
best_final_df_with_only_movie_after_2000s_copy=best_final_df_with_only_movie_after_2000s.copy()
```

```python
def combine_features(row):
    return " ".join([
        str(row['genres']),
        " ".join(row['cast']) if isinstance(row['cast'], list) else str(row['cast']),
        str(row.get('director', ''))
    ])

best_final_df_with_only_movie_after_2000s_copy['combined'] = best_final_df_with_only_movie_after_2000s_copy.apply(combine_features, axis=1)
```

```python
best_final_df_with_only_movie_after_2000s_copy.to_csv('drive/MyDrive/datas/movie_copy_2000s.tsv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=5000, stop_words='english')
X = vectorizer.fit_transform(best_final_df_with_only_movie_after_2000s_copy['combined'])
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=11, metric='cosine', algorithm='brute')
model.fit(X)
```

```python
movie_index = pd.Series(best_final_df_with_only_movie_after_2000s_copy.reset_index().index, index=best_final_df_with_only_movie_after_2000s_copy['title'].str.lower())
```

```python
def recommend(title):
    title = title.lower()
    if title not in movie_index:
        return "Movie not found."

    idx = movie_index[title]
    distances, indices = model.kneighbors(X[idx], n_neighbors=11)

    recommended = []
    for i in indices[0][1:]:
        recommended.append(best_final_df_with_only_movie_after_2000s.iloc[i]['title'])

    return recommended

```

```python
print(recommend("The Avengers"))
```
